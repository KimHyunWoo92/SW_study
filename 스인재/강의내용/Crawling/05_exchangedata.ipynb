# iframe 태그
- 다른 서버의 존재하는 데이터르 현재 페이지에 보여줄 때 사용.
- 원본데이터가 아니여서 접근이 불가능(수정, 삭제 불가)
- 원본데이터가 있는 서버로 접근한 다음에 다시 수집.
- 원본데이터의 주소는 iframe의 src 속성에 담겨져 있다.
- 주소가 절대경로라면(url) 사용 가능 -> 상대경로라면 절대경로로 변형하고 작업
