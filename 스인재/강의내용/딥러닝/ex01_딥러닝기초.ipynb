# Keras 기본 딥러닝 함수
- Sequential() : 신경망을 초기화 하는 함수
- Dense() : 신경망 층을 정의하는 함수 - 퍼셉트론 수, 활성화함수, 입력데이터 크기, 출력데이터 크기등을 설정
- Activation() : 활성화 함수를 정의하는 함수 : sigmoid, Relu등을 사용
- add() :  정의한 신경망층이나 활성화 함수를 신경망에 추가하는 함수
- summary() : 설계한 신경망의 구조를 출력하는 함수
- compile() : 설계한 신경망을 실행할 수 있도록 하는 함수(손실함수, 최적화함수, 평가지표)
- fit() : 설계한 신경망을 이용해서 데이터를 학습하는 함수(특성데이터, 라벨데이터, 배치크기, 반복수, 검증데이터 등을 설정)
- evaluate() : 학습된 모델을 평가하는 함수(오차, 정확도)
- predict() : 학습된 모델을 이용해서 예측하는 함수

# AND 논리 학습

```python
# 데이터 정의 (특성/라벨) - 데이터 수는 4개
X = [[0,0],[0,1],[1,0],[1,1]]
y = [0,0,0,1]
```

```python
import numpy as np
```

```python
X = np.array(X)
y = np.array(y)

X.shape,y.shape
```

# 신경망 설계
- 퍼셉트론 1개로 구성된 신경망

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense,Activation
```

```python
# 신경망초기화
model1 = Sequential()

# 신경망에 신경망층 추가
# units() : 퍼셉트론의 수
model1.add(Dense(units=1,input_dim=2))

# 활성화 함수 추가
model1.add(Activation("sigmoid"))

# 신경망 구조 출력
model1.summary()
```

# 퍼셉트론 2개로 구성된 2층 신경망

```python
# 신경망 초기화
model2 = Sequential()

# 2개층 추가
model2.add(Dense(units=1,input_dim=2))
model2.add(Activation("sigmoid"))
model2.add(Dense(units=1))
model2.add(Activation("sigmoid"))

model2.summary()
```

```python
model3 = Sequential()
model3.add(Dense(units=4,input_dim=2))
model3.add(Activation("sigmoid"))
model3.add(Dense(units=1,input_dim=4))
model3.add(Activation("sigmoid"))

model3.summary()
```

```python
model3 = Sequential()
model3.add(Dense(units=4,input_dim=2))
model3.add(Activation("sigmoid"))
model3.add(Dense(units=1,input_dim=4))
model3.add(Activation("sigmoid"))
```

```python
model1.compile(optimizer="adam",loss="categorical_crossentropy",metrics=["accuracy"])
```

```python
model1.fit(X,y,batch_size=10,epochs=1000)
```

```python
model2.compile(optimizer="adam",loss="categorical_crossentropy",metrics=["accuracy"])
```

```python
h2 = model2.fit(X,y,batch_size=4,epochs=500)
```

```python
model3.compile(optimizer="adam",loss="binary_crossentropy",metrics=["accuracy"])
h3 = model3.fit(X,y,batch_size=4,epochs=500)
```

# 신경망 설계(3층으로 구성)
- 입력층(Input layer)
- 은닉층(Hidden layer)
- 출력층(Output layer)

- 신경망에서 층별 퍼셉트론을 수를 결정하는 예시
    - 인코딩 : 층마다 개수를 점점 줄여가는 방식(특성 선택,압축,요약)
    - 디코딩 : 층마다 개수를 점점 늘려가는 방식(특성 추출,압출해제)

```python
m1 = Sequential()
```

```python
m1.add(Dense(units=16,input_dim=16))
m1.add(Activation("sigmoid"))
m1.add(Dense(units=8))
m1.add(Activation("sigmoid"))
m1.add(Dense(units=1))
m1.add(Activation("sigmoid"))
m1.summary()
```

```python
m1.compile(optimizer="adam",loss="binary_crossentropy",metrics=["accuracy"])
h1 = m1.fit(X,y,batch_size=15,epochs=500)
```

```python
m2 = Sequential()
```

```python
m2.add(Dense(units=8,input_dim=16))
m2.add(Activation("sigmoid"))
m2.add(Dense(units=4))
m2.add(Activation("sigmoid"))
m2.add(Dense(units=1))
m2.add(Activation("sigmoid"))
m2.summary()
```

```python
m2.compile(optimizer="adam",loss="binary_crossentropy",metrics=["accuracy"])
h2 = m2.fit(X,y,batch_size=20,epochs=1000)
```
