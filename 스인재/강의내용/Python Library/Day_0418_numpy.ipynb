#### array 인덱싱 & 슬라이싱
- 인덱싱(indexing) : 값을 가리키다 -> 자료 구조 내에서 1개의 요소에 접근하는 것
- 슬라이싱(slicing) : 값을 잘라오다 -> 자료 구조 내에서 여러개의 요소들에 접근하는 것

#### 2차원 array 인덱싱 슬라이싱
- 인덱싱 : array[행,열]
- 슬라이싱 : array[행시작,:행의끝,열의 시작:열의 끝]

| 함수 이름   | 사용법                 | 설명                                | 예시 결괏값 |
|-------------|------------------------|-------------------------------------|-------------|
| `np.add`    | `np.add(x, y)`         | 배열 x와 y의 요소별 합계            | `3, 7, 11`  |
| `np.subtract` | `np.subtract(x, y)`  | 배열 x에서 y를 요소별로 뺀 결과    | `-1, 1, 3`  |
| `np.multiply` | `np.multiply(x, y)`  | 배열 x와 y의 요소별 곱셈            | `2, 10, 24` |
| `np.divide` | `np.divide(x, y)`      | 배열 x를 y로 요소별 나눈 결과       | `0.5, 2, 4` |
| `np.sqrt`   | `np.sqrt(x)`           | 배열 x의 요소별 제곱근              | `1, 2, 3`   |
| `np.exp`    | `np.exp(x)`            | 배열 x의 요소별 e^x 계산            | `2.718, 7.389, 20.086` |
| `np.log`    | `np.log(x)`            | 배열 x의 요소별 자연로그 계산       | `0, 0.693, 1.099` |
| `np.sin`    | `np.sin(x)`            | 배열 x의 요소별 사인 값 계산        | `0.841, 0.909, 0.141` |
| `np.cos`    | `np.cos(x)`            | 배열 x의 요소별 코사인 값 계산      | `0.540, -0.416, -0.990` |
| `np.tan`    | `np.tan(x)`            | 배열 x의 요소별 탄젠트 값 계산      | `1.557, -2.186, -0.143` |

- 배열 안에서 조건을 충족(특정 조건을 만족)하는 True인 값들만 추출해서 인덱싱하는방법.
- 간단히 말하면, "필터링"을 연상.

#### and,or 논리연산자와 "&, |"같은 비트연산자의 차이
- 논리(and,or) : 단일 Bool값에서만 비교 가능
- 비트(&,|) : 다중 Bool값에 대해서 개별적으로 각 요소에 대한 T/F를 변환

```python
bmi[(bmi >= 23) & (bmi <= 25)]
```

#### 각각의 사용자가 매긴 평점의 평균 중에서 4점 이상의 평점만 구하기.

```python
# 각각의 사용자가 메긴 평점들의 평균 모두 구해보자!
# list에 담아서 for문 활용
# 평점이 4.0 이상인 아이디만 선별

score_avg = []
for i in range(len(np.unique(movie[:,0]))):
    user_id = movie[movie[:,0] == i]
    user_avg = user_id[:,2].mean()
    if user_avg >= 4.0:
        score_avg.append([i,user_avg])


print(score_avg)      
```
