```python
# gruopby, agg 함께 사용
# 제조사별 도시 평균
# 새로운 데이터 프레임을 만드는 개념, agg( 네이밍 컬럼명,(문자열로 존재하는 컬럼명,문자열로 집계함수))
mpg.groupby('제조사').agg(city_mean = ("도시","mean"))
```

```python
# 여러개 사용 가능 agg(1,2,3)
# 제조사별 도시 평균 고속도로 평균
mpg.groupby('제조사').agg(city_mean = ("도시","mean"),high_way =("고속도로연비","mean"))
```

```python
mpg
```

```python
mpg['평균연비'] = mpg[['도시','고속도로연비']].mean(axis=1)
mpg['test결과'] = np.where(mpg['평균연비'] >= 20, 'pass','fail')
```

```python
def makeGrade(row):
    #매개변수가 있는 이유
    #호출될때, row 라는 매개변수에 해당 데이터 프레임을 가지고 오게 된다.

    # row == mpg
    mean = row['평균연비']
    if mean >= 25:
        return "A"
    elif mean >= 20:
        return "B"
    elif mean >= 15:
        return "C"
    else:
        return "F"
```

```python
mpg['grade'] = mpg.apply(makeGrade, axis=1)
```

```python
mpg
```

```python
#그룹별 집단 다시 나누기
# 제조사별 등급 차종의 개수
mpg.groupby('제조사','grade')
```

```python
exam = pd.read_csv("C:\\Users\\SMHRD\\Desktop\\스인재\\강의내용\\data\\exam.csv")
```

```python
exam['total'] = exam[["math","english","science"]].sum(axis=1)
exam['mean'] = exam[["math","english","science"]].mean(axis=1)
exam['result'] = np.where(exam['mean']>=60,"합격","불합격")
```

```python
def grade(row):
    target = row['mean']
    if target >=80:
        return "1등급"
    elif target >= 70:
        return "2등급"
    elif target >= 60:
        return "3등급"
    else:
        return "재시험"
```

```python
exam['Grade'] = exam.apply(grade,axis=1)
```

```python
exam
```

```python
# 반별 합격/불합격자 수
ex_pass = exam.query("result in ['합격']")
ex_pass.groupby('nclass').agg(pass_result=("result","count"))
```

```python
ex_fail = exam.query("result in ['불합격']")
ex_fail.groupby('nclass').agg(fail_result=("result","count"))        
```

```python
exam.groupby(['nclass','result']).agg(ck_result=('result','count')).sort_index(ascending=(True,False))
```

```python
# 반별 등급의 수
exam.groupby(['nclass','Grade']).agg(grade_count=("Grade","count"))
```
