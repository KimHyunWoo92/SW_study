## 데이터 시각화
- 시각화를 통해 데이터의 특징을 쉽게 파악
- 분석 결과를 상대방에게 효과적으로 전달 가능

## Matplotlib + "Seaborn"
- python에서 시각화 할 수 있는 대표적인 라이브러리
- pyplot,pylab이란 서브 패키지 존재
- 간단하게 정보를 입력해서 그래프를 그려주는 pyplot 사용

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# line plot
x = np.arange(1,6)
np.random.seed(3)
y = np.random.randint(1,10,size=5)
print(y)

#그리는 법
# plt.plot(x,y)
plt.plot(x,y)
plt.show()
```

```python
#plot의 경우 x축 생략 가능
plt.plot(y)
plt.show()
```

```python
def test1(num1=10):
    print(num1)
```

```python
#스타일 바꾸기
#함수에는 매개변수 기본으로 줄 수 있다.
plt.plot(x,y,ls=':',c='red',marker='*',ms=15,mew=5,mfc='c')
plt.show()
```

```python
#한 셀에 4개의 그래프 그래기
plt.plot(x,y, ls='-',c='b')

plt.plot(x,y, ls='--',c='r')

plt.plot(x,y, ls='-.',c='y')

plt.plot(x,y, ls=':',c='m')
# ㅎplot을 여러개 작성 할 경우 하나의 그래프에 겹쳐서 나온다.
```

```python
#subplot(row, column, index)
#2,2로 진행
plt.subplot(2,2,1)
plt.plot(x,y, ls='-',c='b')

plt.subplot(2,2,2)
plt.plot(x,y, ls='--',c='r')

plt.subplot(2,2,3)
plt.plot(x,y, ls='-.',c='y')

plt.subplot(2,2,4)
plt.plot(x,y, ls=':',c='m')
```

```python
#타이틀주기
x1 = np.arange(1,6)
np.random.seed(4)

y1 = np.random.randint(1,10,size=5)

plt.title("test")

plt.plot(x1,y1)
plt.show()
```

```python
#grid
plt.grid()

plt.plot(x1,y1)
plt.show()
```

```python
plt.plot(x,y,label='test1')
plt.plot(x1,y1,label='test2')
plt.legend(loc='upper right')
plt.show()
```

### xlabel, ylabel



```python
plt.plot(x,y,label='test1')
plt.plot(x1,y1,label='test2')

plt.xlabel('x')
plt.ylabel('y')

plt.legend(loc=1)
plt.show()
```

### xlim, ylim

```python
# xlim, ylim

plt.plot(x,y,label='test1')
plt.plot(x1,y1,label='test2')

plt.xlabel('x')
plt.ylabel('y', rotation='horizontal')

plt.legend(loc=1)

# 보다 넓게
plt.xlim(1,10)

# 보다 좁게
plt.ylim(1,5)
plt.show()
```
