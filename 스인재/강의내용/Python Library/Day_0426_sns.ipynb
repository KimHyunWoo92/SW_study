```python
df = pd.read_csv("../data/202311_202311_연령별인구현황_월간.csv",encoding='euc-kr')
df
```

```python
pop = df.copy()
pop
```

```python
pop.query("행정구역 == '광주광역시 서구 금호1동(2914082100)'")
```

```python
pop_kumho = pop.query("행정구역 == '광주광역시 서구 금호1동(2914082100)'").set_index('행정구역')
pop_kumho
```

```python
pop_kumho.iloc[:,2:103].columns
```

```python
pop_male = pop_kumho.iloc[:,105:206]
pop_male
```

```python
pop_female = pop_kumho.iloc[:,208:309]
pop_female
```

```python
male_arr = np.array(pop_male).flatten()
female_arr = np.array(pop_female).flatten()

plt.plot(male_arr,label="male")

plt.plot(female_arr,label="female")
plt.legend()
plt.show()
```

```python
male_list = []
female_list = []

for i in male_arr:
    male_list.append(int(i))
for i in female_arr:
    female_list.append(-int(i))
```

```python
#barh
plt.style.use('ggplot')
plt.barh(range(101), male_list, label='male')
plt.barh(range(101), female_list, label='female')

plt.legend()
plt.show()
```
