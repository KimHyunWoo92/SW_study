```python
# 파생변수 : 조건에 따라 서로 다른 값
# 평균연비 20이상이면 pass 아니면 fail
```

```python
# 사용방법 np.where(조건식,True인경우, False인경우)
mpg['test결과'] = np.where(mpg['평균연비'] >= 20, 'pass','fail')
mpg
```

```python
# 조금 더 복잡한 조건식을 사용하여 파생변수
# grade 라는 컬럼
# 평균연비 25이상 A, 20이상 B, 15이상 C,그 미만 F

np.where(mpg['평균연비'] >= 25,'A',np.where(mpg['평균연비']>=20,"B"),mpg['평균연비'])
```

```python
def makeGrade(row):
    #매개변수가 있는 이유
    #호출될때, row 라는 매개변수에 해당 데이터 프레임을 가지고 오게 된다.

    # row == mpg
    mean = row['평균연비']
    if mean >= 25:
        return "A"
    elif mean >= 20:
        return "B"
    elif mean >= 15:
        return "C"
    else:
        return "F"
```

```python
#apply()
# 매개변수로 함수
# apply(함수명, axis=1)
mpg.apply(makeGrade, axis=1)
```

```python
#데이터 불러오기
df1 = pd.read_csv("C:\\Users\\SMHRD\\Desktop\\스인재\\강의내용\\data\\exam.csv")
exam = df1.copy()
```

```python
# 데이터 정보 보기
exam.info()
```

```python
# 데이터 요약통계 보기
exam.describe()
```

```python
#모든 과목의 합을 파생변수로 만들기(컬럼명은 total)
exam['total'] = exam[["math","english","science"]].sum(axis=1)
```

```python
exam
```

```python
#모든 과목의 평균 파생변수로 만들기(컬럼명은 mean)
exam['mean'] = exam.loc[:100,"math":"science"].mean(axis=1).round(2)
```

```python
exam
```

```python
# 평균이 60점 이상이면 합격, 아니면 불합격 파생면수 만들기 (컬럼명 result)
exam['result'] = np.where(exam['mean']>=60,"합격","불합격")
```

```python
#result가 합격/불합격 수 알아보기!
#exam['컬럼명'].value_counts()
exam['result'].value_counts()
```

```python
def grade(row):
    target = row['mean']
    if target >=80:
        return "1등급"
    elif target >= 70:
        return "2등급"
    elif target >= 60:
        return "3등급"
    else:
        return "재시험"
```

```python
# 평균이 80점 이상이면 1등급,70점 이상이면 2등급, 60점 이상이면 3등급 아니면 재시험 파생변수 만들기(컬럼명 grade)
exam['Grade'] = exam.apply(grade,axis=1)
```

```python
exam['Grade'].value_counts().sort_index()
```
